```python
import pandas as pd
import numpy as np

data=pd.read_csv("train.csv")
```

```python
X_train=data.drop(columns=["Cover_Type"])
y_train=data['Cover_Type']
```

```python
X_test=pd.read_csv("test.csv")
X_test
```

```python
#Supprimer les colonnes d'ID
X_test=X_test.drop(columns=["row_ID"])
X_train=X_train.drop(columns=['Unnamed: 0'])
```

```python
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

from joblib import load, dump
```

```python
rf = RandomForestClassifier(n_jobs=-1)
rf.fit(X_train,y_train)
```

```python
X_train.columns
```

```python
X_test.columns
```

```python
y_pred = rf.predict(X_test)
```

```python
submission_df = pd.DataFrame({
    'row_ID': X_test.index,   
    'Cover_Type': y_pred      
})

submission_df.to_csv('naive_submission_1.csv', index=False)
```

```python
submission_df
```

```python

#Sauvegarder le modèle
dump(rf, 'rf.joblib')
```

```python
features_scores=pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
features_scores
```
